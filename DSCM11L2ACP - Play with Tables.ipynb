```python
import sqlite3

database = 'basketball.sqlite'

conn = sqlite3.connect(database)

print("Opened Database Successfully...")
```

```python
import pandas as pd
 
tables = pd.read_sql("""SELECT *
                    FROM sqlite_master
                    WHERE type=='table';""", conn)
```

```python
tables
```

```python
result = pd.read_sql("""SELECT * 
                    FROM Team;""", conn)
result
```

```python
result = pd.read_sql("""SELECT full_name, nickname, city, year_founded 
                    FROM Team
                    WHERE year_founded > 1990;""", conn)
result
```

```python
result = pd.read_sql("""SELECT *
                    FROM Team
                    WHERE state IN ('New York', 'Texas');""", conn)
result
```

```python
result = pd.read_sql("""SELECT * 
                    FROM Team
                    WHERE full_name LIKE 'L%';""", conn)
result
```

```python
result = pd.read_sql("""SELECT MIN(year_founded), MAX(year_founded) 
                    FROM Team;""", conn)
result
```

```python
result = pd.read_sql("""SELECT * 
                    FROM Team
                    WHERE year_founded IN (1946, 2002);""", conn)
result
```

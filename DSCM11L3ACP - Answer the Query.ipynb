```python
# Connect with sqlite database
# Import necessary libraries
import sqlite3

database = 'basketball.sqlite'

conn = sqlite3.connect(database)
print('Opened data successfully')
```

```python
import pandas as pd
 
tables = pd.read_sql("""SELECT *
                    FROM sqlite_master
                    WHERE type=='table';""", conn)

tables
```

```python
result = pd.read_sql("""SELECT * 
                    FROM Draft;""", conn)
result
```

```python
result = pd.read_sql("""SELECT nameTeam, COUNT(DISTINCT idPlayer) 
                    FROM Draft
                    GROUP BY nameTeam
                    ORDER BY COUNT(DISTINCT idPlayer) DESC;""", conn)
result.head(10)
```
